```python
import pandas as pd
import numpy as np

# Load datasets
df1 = pd.read_csv("../data/raw/mutual_fund.csv")
df2 = pd.read_csv("../data/raw/hdfc_top100.csv")

# -------------------------------
# 1. Basic Info
# -------------------------------
print("DF1 Shape:", df1.shape)
print("DF2 Shape:", df2.shape)

print("\nDF1 Info:")
print(df1.info())

print("\nDF2 Info:")
print(df2.info())

# -------------------------------
# 2. Missing Values
# -------------------------------
print("\nMissing values DF1:\n", df1.isnull().sum())
print("\nMissing values DF2:\n", df2.isnull().sum())

# -------------------------------
# 3. Remove Duplicates
# -------------------------------
df1 = df1.drop_duplicates()
df2 = df2.drop_duplicates()

# -------------------------------
# 4. Convert Date Column
# -------------------------------
df2['date'] = pd.to_datetime(df2['date'], dayfirst=True)
df2 = df2.sort_values(by='date')

# -------------------------------
# 5. Handle Missing Values
# -------------------------------
df1 = df1.ffill()
df2 = df2.ffill()

# -------------------------------
# 6. Final Check
# -------------------------------
print("\nAfter Cleaning:\n")
print(df1.isnull().sum())
print(df2.isnull().sum())

# -------------------------------
# 7. Save Cleaned Data
# -------------------------------
df1.to_csv("../data/processed/mutual_fund_cleaned.csv", index=False)
df2.to_csv("../data/processed/hdfc_top100_cleaned.csv", index=False)

print("✅ Data Cleaning Completed Successfully!")
# Save cleaned files

df1.to_csv("../data/processed/mutual_fund_cleaned.csv", index=False)
df2.to_csv("../data/processed/hdfc_top100_cleaned.csv", index=False)

print("✅ Cleaned files saved to processed folder!")
```
